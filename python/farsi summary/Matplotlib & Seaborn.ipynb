```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, 'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title Here')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y, 'b')
axes2.plot(y,x, 'r')
```

## ضریب همبستگی:
میزان ارتباط خطی بین دو دیتاست رو اندازه میگیره و پیشنیازش اینه که دیتاستها هر کدام به صورت توزیع نرمال در اومده باشند. مثل همه ضرایب همبستگی، بین -1 و +1 متغیره و 0 به معنی عدم همبستگی هست. همچنین -1 و +1 به معنای وجود ارتباط دقیق خطیه. و هر عدد مثبت دیگه ای به معنی اینه که اگر x زیاد بشه، y هم زیاد میشه. و تو عددای منفی یعنی اگر x کم بشه y هم کم میشه.
یه نکته دیگه اینکه این عدد خیلی قابل اتکا نیست و برای دیتاست های بزرگتر از 500 تازه منطقی در میاد.
